# Covid-19 Worldwide cases Analysis 

**Importing libraries**

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

**Will check Data for analysis using pandas library**

```python
covid = pd.read_csv('/kaggle/input/corona-virus-report/country_wise_latest.csv')
covid
```

**Checking top 10 countries with most deaths.**

```python
covid.nlargest(10,'Deaths')   
```

**Sameway for the less deaths.**

```python
covid.nsmallest(10,'Deaths')
```

**now checking null values in the dataset.**

```python
covid.isnull().sum()
```

**country with max deaths**

```python
covid.max()
```

**sameway country with least deaths.**

```python
covid.min()
```

**Describing Data**

```python
covid.describe()
```

**lets visualize the top 20 countriees**

```python
plt.figure(figsize=(10,15))
covid['Country/Region'].head(20).value_counts().plot.pie(autopct="%1.1f%%")     #Note use of pie diagram.
plt.title('Top 20 Country With Max Cases',fontsize=9)                         #Note Labelling of diagram
plt.ylabel('Countries',fontsize=10)
plt.show()
```

**visualizing top confirmed cases by country**

```python
plt.style.use('Solarize_Light2')
plt.figure(figsize=(15,10))
sns.countplot(data=covid.head(20),y='Confirmed')                    #Note here we are using countplot as sns which is an in-nuilt library in python that uses mtplotlib underneath to plot graphs.
plt.xticks(rotation='vertical')                                    #Note use of ticks for a better view of diagram.
plt.title('Top Confirmed Cases By Country',fontsize=25)            #Note labelling the data
plt.xlabel('Confirmed Cases')                               
plt.show()
```

**now visualizing top 20 record countries**

```python
plt.style.use('Solarize_Light2')
plt.figure(figsize=(15,10))
sns.countplot(data=covid.head(20),y='Confirmed')                    #Note here we are using countplot as sns which is an in-nuilt library in python that uses mtplotlib underneath to plot graphs.
plt.xticks(rotation='vertical')                                    #Note use of ticks for a better view of diagram.
plt.title('Top Confirmed Cases By Country',fontsize=25)            #Note labelling the data
plt.xlabel('Confirmed Cases')                               
plt.show()
```

**top 20 Deaths by countries**

```python
plt.figure(figsize=(10,15))       #Note the figure size
covid['Deaths'].head(20).value_counts().plot.pie(autopct="%1.1f%%")    #Note use of pie plot
plt.title('Top 20 Deaths With Countries',fontsize=25)                 #Note leabelling the diagram
plt.xlabel('Countries',fontsize=15)
plt.show()
```
